## Regression

```python
def prepare_boston():
    boston = load_boston()
    X = pd.DataFrame(data=boston.data, columns=boston.feature_names)
    y = pd.Series(data=boston.target, name='price')
    
    return X, y


X_, y_ = prepare_boston()
y_
```

```python
dtc = DecisionTreeRegressor()
dtc.fit(X_, y_)
dtc.predict(X_)
```

# Random Forests

## Classification

```python
dtc = RandomForestClassifier()
dtc.fit(X, y)
dtc.predict(X)
```

## Regression

```python
dtc = RandomForestRegressor()
dtc.fit(X_, y_)
dtc.predict(X_)
```

# Gradient Boosting
